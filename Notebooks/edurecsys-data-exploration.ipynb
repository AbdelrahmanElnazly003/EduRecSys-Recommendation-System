# EduRecSys – Data Exploration

This notebook performs exploratory data analysis (EDA) on the datasets used in the EduRecSys project.

The goal of this notebook is to:
- Understand the structure of the Udemy courses dataset (English).
- Explore the SANAD Arabic dataset (Tech and Finance categories).
- Validate data quality, balance, and text characteristics.

```python
import os
import pandas as pd
from pathlib import Path

BASE_PATH = Path("/kaggle/input/edurecsys-raw-data")

UDEMY_PATH = BASE_PATH / "udemy_courses.csv"
SANAD_PATH = BASE_PATH / "/kaggle/input/edurecsys-raw-data/sanad/sanad"
```

```python
# Load Udemy Dataset
udemy_df = pd.read_csv(UDEMY_PATH)

print("Udemy shape:", udemy_df.shape)
udemy_df.head()
```

# Udemy Dataset Observations
- Udemy dataset provides structured course-level metadata.
- Subjects are aligned with professional educational domains.
- Course levels are well distributed, with a focus on beginner and all-level audiences.
- Dataset is suitable for building content-based and hybrid recommendation systems.


```python
# Load SANAD (Tech + Finance)
sanad_records = []

for subject in ["Tech", "Finance"]:
    subject_path = SANAD_PATH / subject
    
    for file_name in os.listdir(subject_path):
        if file_name.endswith(".txt"):
            file_path = subject_path / file_name
            
            with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                text = f.read().strip()
            
            sanad_records.append({
                "content_id": f"{subject}_{file_name.replace('.txt','')}",
                "description": text,
                "subject": subject
            })

sanad_df = pd.DataFrame(sanad_records)

print("SANAD shape:", sanad_df.shape)
sanad_df.head()
```

```python
sanad_df["subject"].value_counts()
```

```python
# Text length
sanad_df["text_length"] = sanad_df["description"].apply(lambda x: len(x.split()))
sanad_df["text_length"].describe()
```

# SANAD Dataset Observations
- SANAD dataset contains Arabic articles from Tech and Finance domains.
- The dataset is balanced across selected categories.
- Article lengths are sufficient for NLP-based similarity modeling.
- Selected domains align well with professional educational content.
